## Drop irrelevant rows, clean names and drop omitted rows

```python
reg = reg.dropna()
```

```python
reg = reg[6:-8].copy()
```

```python
reg['Unnamed: 0'] = reg['Unnamed: 0'].replace('Name = \d+,',' ',regex=True).astype(str)
```

```python
reg = reg.rename(columns = {'Unnamed: 0': 'Name', '(1)': 'Agentic', '(2)': 'Communal'}).copy()
```

```python
reg = reg[reg['Name'].str.contains('omitted') == False].copy()
```

```python
reg = reg.sort_values('Name').copy()
```

```python
reg = reg.reset_index(drop = True)
```

## concat

```python
new = pd.concat([Gender, reg], axis = 1)
```

```python
new 
```

## Cleaning Numbers

```python
new = new.drop(columns = ['Name '])
```

```python
new['Agentic_Coef'] = new[['Agentic']].replace('\*','',regex=True).astype(float)
```

```python
new['Communal_Coef'] = new[['Communal']].replace('\*','',regex=True).astype(float)
```

## Normalizing 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
new['Agentic_Coef_N'] = scaler.fit_transform(new[['Agentic_Coef']]).round(2)
```

```python
new['Communal_Coef_N'] = scaler.fit_transform(new[['Communal_Coef']]).round(2)
```

```python
new
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
Female = new[new['Female'] == 1]
sns.distplot(Female['Agentic_Coef_N'],  kde=False, label='Female')
Male = new[new['Female'] == 0]
sns.distplot(Male['Agentic_Coef_N'],  kde=False, label='Male')
plt.legend(prop={'size': 12})
plt.title('Normalized Agentic Model Coefficients with Individual and Source Fixed Effects')
plt.xlabel('Agentic_Coef_N')
plt.ylabel('Frequency')
```

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot("Female", "Agentic_Coef_N", "Female", data=new, kind="box")
    g.set_axis_labels("Female", "Agentic_Coef_N");
```

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot("Female", "Communal_Coef_N", "Female", data=new, kind="box")
    g.set_axis_labels("Female", "Communal_Coef_N");
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
Female = new[new['Female'] == 1]
sns.distplot(Female['Communal_Coef_N'],  kde=False, label='Female')
Male = new[new['Female'] == 0]
sns.distplot(Male['Communal_Coef_N'],  kde=False, label='Male')
plt.legend(prop={'size': 12})
plt.title('Normalized Communal Model Coefficients with Individual and Source Fixed Effects')
plt.xlabel('Communal_Coef_N')
plt.ylabel('Frequency')
```
