```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
forest_cv_score = cross_val_score(forest, X, y, cv=10, scoring= "roc_auc")
```

```python
print("=== Confusion Matrix ===")
print(confusion_matrix(y_Test, predictions_forest))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_Test, predictions_forest))
print('\n')
print("=== All AUC Scores ===")
print(forest_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", forest_cv_score.mean())
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
# number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# number of features at every split
max_features = ['auto', 'sqrt']

# max depth
max_depth = [int(x) for x in np.linspace(100, 500, num = 11)]
max_depth.append(None)
# create random grid
random_grid = {
 'n_estimators': n_estimators,
 'max_features': max_features,
 'max_depth': max_depth
 }
# Random search of parameters
forest_random = RandomizedSearchCV(estimator = forest, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, n_jobs = -1)
# Fit the model
forest_random.fit(X_Train, y_Train)
# print results
print(forest_random.best_params_)
```

```python
forest_tuned = RandomForestClassifier(n_estimators=800, max_depth=140, max_features='auto')
forest_tuned.fit(X_Train,y_Train)
forest_tuned_predict = forest_tuned.predict(X_Test)
forest_tuned_cv_score = cross_val_score(forest_tuned, X, y, cv=10, scoring='roc_auc')
print("=== Confusion Matrix ===")
print(confusion_matrix(y_Test, forest_tuned_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_Test, forest_tuned_predict))
print('\n')
print("=== All AUC Scores ===")
print(forest_tuned_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", forest_tuned_cv_score.mean())
```
